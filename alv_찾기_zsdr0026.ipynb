```python
# -*- coding: utf-8 -*-
"""
find_alv_zsdr0026.py
──────────────────────────────────────────
ZSDR0026 F8 결과 화면을 열어둔 상태에서 실행.d
SAP GUI 전체 컨트롤 트리를 덤프하여 ALV Grid 경로를 확인.
결과: alv_tree_zsdr0026.txt
"""

import win32com.client
import sys

OUTPUT_FILE = "alv_tree_zsdr0026.txt"


def dump_tree(obj, depth=0, out_lines=None):
    if out_lines is None:
        out_lines = []
    try:
        obj_type = obj.Type
    except Exception:
        obj_type = "Unknown"
    try:
        obj_id = obj.Id
    except Exception:
        obj_id = "?"

    extra = ""
    if obj_type in ("GuiShell", "GuiGridView"):
        try:
            extra += f"  RowCount={obj.RowCount}"
        except Exception:
            pass
        try:
            extra += f"  ColCount={obj.ColumnCount}"
        except Exception:
            pass
        try:
            extra += f"  SubType={obj.SubType}"
        except Exception:
            pass

    line = f"{'  ' * depth}[{obj_type}]  {obj_id}{extra}"
    out_lines.append(line)
    print(line)

    try:
        count = obj.Children.Count
    except Exception:
        return out_lines

    for i in range(count):
        try:
            child = obj.Children.ElementAt(i)
            dump_tree(child, depth + 1, out_lines)
        except Exception as e:
            out_lines.append(f"{'  ' * (depth+1)}[ERROR] {e}")

    return out_lines


def main():
    try:
        sap_gui = win32com.client.GetObject("SAPGUI")
        app     = sap_gui.GetScriptingEngine
        conn    = app.Children(0)
        session = conn.Children(0)
    except Exception as e:
        print(f"SAP 연결 실패: {e}")
        sys.exit(1)

    wnd = session.findById("wnd[0]")
    lines = dump_tree(wnd)

    with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
        f.write("\n".join(lines))

    print(f"\n✅ 저장 완료: {OUTPUT_FILE}")
    print("이 파일에서 [GuiShell] 또는 [GuiGridView] 항목의 경로를 확인하세요.")


if __name__ == "__main__":
    main()
```
